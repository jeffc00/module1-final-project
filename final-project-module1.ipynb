- From **B3** website, using *Selenium library*

```python
url = 'http://www.b3.com.br/pt_br/market-data-e-indices/servicos-de-dados/market-data/consultas/mercado-a-vista/volatilidades-dos-ativos/'

# initializing the webdriver
driver = webdriver.Chrome('../../chromedriver.exe')
```

```python
periods = ['1 mes', '3 meses', '6 meses', '1 ano']
volat_b3 = {}

for i, per in enumerate(periods):
    driver.get(url)
    
    time.sleep(10)
    
    frame_elem = driver.find_element_by_id('bvmf_iframe')
    driver.switch_to.frame(frame_elem)
    driver.find_element_by_id(f'ctl00_contentPlaceHolderConteudo_rblPeriodo_rblPeriodo_{i}').click()
    driver.find_element_by_id('ctl00_contentPlaceHolderConteudo_btnTodos').click()
    print(per)

    time.sleep(10)
    
    driver.find_element_by_id('divContainerIframe')
    header_ls = (driver.find_element_by_id('ctl00_contentPlaceHolderConteudo_grdVolatilidade_ctl01')
                 .find_element_by_tag_name('thead').find_elements_by_tag_name('th'))
    volat_b3[per] = [[th.text for th in header_ls]]

    for i in range(16):

        print(i + 1)

        body_ls = (driver.find_element_by_id('ctl00_contentPlaceHolderConteudo_grdVolatilidade_ctl01')
                   .find_elements_by_tag_name('tbody')[1].find_elements_by_tag_name('tr'))
        td_body_ls = [list(map(lambda td: td.text, tr.find_elements_by_tag_name('td'))) for tr in body_ls]
        
        for i, td in enumerate(td_body_ls):
            if td[0] not in list(tickers_ibov):
                continue
            else:
                volat_b3[per].append(td_body_ls[i])
                
        time.sleep(10)

        try:
            (driver.find_element_by_id('ctl00_contentPlaceHolderConteudo_grdVolatilidade_ctl01_ctl03_ctl01_ctl00_lnkNext')
             .click())
        except:
            break
        else:
            print('change page')
            time.sleep(10)
```

```python
volat_b3_df = {}
for per in periods:
    volat_b3_df[per] = pd.DataFrame(volat_b3[per], columns=volat_b3[per][0]).drop(0, axis=0).reset_index(drop=True)
    volat_b3_df[per].drop(['Nome', 'Especificação', 'Desvio Padrão'], axis=1, inplace=True)
    volat_b3_df[per].columns = ['ticker', per]
    volat_b3_df[per][per] = volat_b3_df[per][per].apply(lambda x: x + '%')
    
volat_b3_df = (volat_b3_df['1 mes']
               .merge(volat_b3_df['3 meses'], on='ticker')
               .merge(volat_b3_df['6 meses'], on='ticker')
               .merge(volat_b3_df['1 ano'], on='ticker')
               .drop(range(57, 72), axis=0)
               .reset_index(drop=True)
              )
# volat_b3_df
```

# 1.3 Extracting from Alpha Vantage API

### 1.3.1 Daily time series

```python
apikey = 'Q3QSDCKLVDZPE32H'
url = 'https://www.alphavantage.co/query?'
```

```python
# create a function called "chunks" which takes a list l and breaks it up into chunks of size at most n
# this will be useful later on when making requests to the API
def chunks(l, n):
    for i in range(0, len(l), n):
        yield l[i:i+n]
```

```python
# monthly time series of the BOVESPA stocks
function = 'TIME_SERIES_MONTHLY'
outputsize = 'full'
ticker = ibov['ticker'].apply(lambda t: t + '.SA').append(pd.Series(['^BVSP']), ignore_index=True)

for chunk in chunks(tickers, 5):
    for ticker in chunk:
        response = requests.get(url + '&'.join([f'function={function}', f'symbol={ticker}',\
                                                f'outputsize={outputsize}', f'apikey={apikey}']))
        print(response)
        result = response.json()
        result_df = pd.DataFrame.from_dict(result['Monthly Time Series'], orient='index')
        result_df.columns = ['open', 'high', 'low', 'close', 'volume']
        result_df.to_csv(f'output/{ticker}.csv')
        
    time.sleep(62)
```

## 2.3 Correlation heatmap of the Ibovespa's stocks historical monthly returns

```python
files = os.listdir(path='output/')
index = pd.to_datetime(pd.read_csv('output/^BVSP.csv', index_col=0).index, infer_datetime_format=True)
data = pd.DataFrame(index=index)
for f in files:
    returns_df[f[:5]] = pd.read_csv(f'output/{f}', index_col=0)
    returns_df[f[:5]].index = pd.to_datetime(returns_df[f[:5]].index, infer_datetime_format=True)
    returns_df[f[:5]]['monthly_return'] = (returns_df[f[:5]]['close'] / returns_df[f[:5]]['open'] - 1) * 100
    returns_df[f[:5]] = returns_df[f[:5]][['monthly_return']]
    returns_df[f[:5]].columns = [f[:5]]
    data = data.join(returns_df[f[:5]])
```

```python
corr_df = data.corr()

fig, ax = plt.subplots(figsize=(16, 16))

mask = np.zeros_like(corr_df, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True
sns.heatmap(corr_df, mask=mask, cmap='Blues', square=True, cbar_kws={"shrink": .75}, linewidths=0.1, ax=ax)
ax.set_ylim(len(corr_df), 0, 0)
ax.set_title('Mapa de Correlação dos Retornos Mensais', size=20, pad=30);
```
